# **Introduction** <a class="anchor" id="0"></a>
"Pendidikan adalah fondasi utama bagi kemajuan suatu bangsa. Di Indonesia, kualitas pendidikan masih menghadapi berbagai tantangan yang memerlukan perhatian serius. Oleh karena itu, penting untuk menganalisis faktor-faktor yang memengaruhi kualitas pendidikan di Indonesia dari tahun 2014 hingga 2023.

Penelitian sebelumnya menunjukkan bahwa faktor-faktor seperti infrastruktur pendidikan, kualitas guru, dan partisipasi masyarakat sangat memengaruhi kualitas pendidikan di berbagai daerah (Purnastuti et al., 2019; Sukarno, 2020). Selain itu, kebijakan pendidikan yang diterapkan pemerintah juga memiliki dampak signifikan terhadap hasil pendidikan (Fajri & Rachmawati, 2021).

Dengan menganalisis tingkat partisipasi pendidikan dan faktor-faktor yang memengaruhinya, kita dapat memperoleh gambaran yang lebih jelas tentang kondisi pendidikan di Indonesia. Penelitian ini tidak hanya bertujuan untuk memahami kondisi saat ini tetapi juga untuk mengidentifikasi kebijakan yang tepat guna meningkatkan kualitas pendidikan di masa depan.

Urgensi dari penelitian ini terletak pada kebutuhan untuk memastikan bahwa kebijakan pendidikan yang diterapkan saat ini masih relevan dan efektif. Jika tidak, diperlukan rekomendasi kebijakan baru yang lebih sesuai dengan perkembangan zaman dan kebutuhan masyarakat. Analisis ini diharapkan dapat memberikan arahan yang jelas bagi pembuat kebijakan dalam merumuskan strategi yang tepat untuk meningkatkan kualitas pendidikan di Indonesia."

<li> Indonesia telah mengalami beberapa reformasi kurikulum untuk meningkatkan kualitas pendidikan dan membuatnya lebih relevan dengan kebutuhan abad ke-21. ## 2.2 **Rasio Jumlah Guru dan Siswa** <a class="anchor" id="2.2"></a>

```python
combined_df = pd.concat([df_rasio_sma, df_rasio_smk, df_rasio_ma, df_rasio_smp, df_rasio_mts, df_rasio_mi])

data_rasio = combined_df.groupby('provinsi').agg({'jumlah guru (negeri+swasta)': 'sum', 'jumlah murid (negeri+swasta)': 'sum'}).reset_index()

data_rasio.head()
```

```python
data_rasio['rasio'] = data_rasio['jumlah guru (negeri+swasta)'] / data_rasio['jumlah murid (negeri+swasta)']

data_rasio = data_rasio[['provinsi', 'rasio']].sort_values(by='provinsi')

data_rasio = data_rasio.reset_index(drop=True)
```

```python
data_rasio.head()
```

```python
data_rasio_sorted = data_rasio.sort_values(by='rasio', ascending=True)
data_rasio_sorted = data_rasio_sorted.reset_index(drop=True)

data_rasio_sorted.head()
```

### Analisis Hubungan Rasio Guru dan Siswa dengan Indeks Pendidikan di Provinsi-Provinsi di Indonesia

Gambar di atas menunjukkan perbandingan antara rasio guru terhadap siswa dan indeks pendidikan di berbagai provinsi di Indonesia.

**Rasio Guru-Siswa yang Tinggi:**
- Provinsi dengan rasio guru terhadap siswa yang tinggi, seperti DI Yogyakarta dan Kalimantan Utara, cenderung memiliki indeks pendidikan yang lebih baik. Hal ini menunjukkan bahwa lebih banyak guru per siswa dapat berdampak positif pada kualitas pendidikan.

**Rasio Guru-Siswa yang Rendah:**
- Provinsi seperti Papua Barat dan Nusa Tenggara Timur, yang memiliki rasio guru terhadap siswa yang rendah, juga menunjukkan indeks pendidikan yang lebih rendah. Kekurangan guru dapat mengakibatkan kurangnya perhatian individu terhadap siswa, yang pada akhirnya mempengaruhi kualitas pendidikan.

**Kesenjangan Antar Provinsi:**
- Terdapat kesenjangan yang cukup signifikan antara provinsi-provinsi dalam hal rasio guru terhadap siswa dan indeks pendidikan. Misalnya, provinsi di Jawa umumnya memiliki indeks pendidikan yang lebih tinggi dibandingkan dengan provinsi di bagian timur Indonesia.
